```python
## 통계함수

import pandas as pd

# read_csv() 함수로 df 생성
df = pd.read_csv('./data/auto-mpg.csv')

# 열 이름을 지정
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower',	'weight', 'acceleration', 'model year', 'origin', 'name']

# 평균값
print(df.mean(numeric_only=True)) #numeric_only=True: 산술 데이터를 가진 7개 열에 대한 평균값만 계산
print('\n')
print(df['mpg'].mean())
print(df.mpg.mean())
print('\n')
print(df[['mpg', 'weight']].mean())
```

```python
# 표준편차
print(df.std(numeric_only=True)) #numeric_only=True: 산술 데이터를 가진 7개 열에 대한 평균값만 계산
print('\n')
print(df['mpg'].std())
```

```python
# 상관계수
print(df.corr(numeric_only=True)) #numeric_only=True: 산술 데이터를 가진 7개 열에 대한 평균값만 계산
print('\n')
print(df[['mpg', 'weight']].corr())
```

```python
## 미션1
# 데이터 전처리 및 결측치 처리하기
# 미션 1-1: horsepower 열의 결측치가 있는지 확인하고, 결측치를 horsepower의 중앙값(median)으로 대체하세요.
import pandas as pd

# read_csv() 함수로 df 생성
df = pd.read_csv('./data/auto-mpg.csv')

# 열 이름을 지정
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower',	'weight', 'acceleration', 'model year', 'origin', 'name']

df.dtypes
```

```python
# df = df.astype(float)로 하는 경우 ? 오류뜸 -> ?을 NaN 값으로 바꿔주기 위해 다음과 같이 처리함
# horsepower 열의 문자형 -> 숫자형으로 변경
df2 = df.copy()
df2['horsepower'] = pd.to_numeric(df2['horsepower'], errors='coerce')
df2.dtypes
```

```python
# horsepower 열의 결측치가 있는지 확인
print(df2['horsepower'].isnull().sum())
print('\n')
# horsepower 열의 결측치 직접 확인
print(df2[df2['horsepower'].isnull()])
print('\n')
# horsepower 중앙값 구하기
median_value = df2['horsepower'].median()
print(median_value)
```

```python
# 결측치를 horsepower의 중앙값(median)으로 대체
print(df2['horsepower'].fillna(df2['horsepower'].median(), inplace=True))

# 결측치 없는지 확인
assert df2['horsepower'].isnull().sum() == 0

# 중앙값으로 채워졌는지 눈으로도 확인
print(df2['horsepower'].value_counts().sort_index())
```

```python
## 미션2
# 조건부 데이터 필터링 및 새로운 열 생성
# 미션 2-1: 연비(mpg)가 30 이상인 차량만 필터링하여 high_mpg라는 새로운 DataFrame을 생성하세요.

high_mpg = df2[df2['mpg']>=30]
high_mpg
```

```python
# 미션 2-2: high_mpg 데이터에 무게(weight) 대비 마력(horsepower)의 비율을 나타내는 hp_weight_ratio라는 새로운 열을 추가하세요.

high_mpg['hp_weight_ratio'] = high_mpg['horsepower'] / high_mpg['weight'] 
high_mpg
```

```python
## 미션3
# 그룹화 및 통계 분석 심화
# cylinders와 origin을 기준으로 그룹화하여, 그룹별로 다음을 계산하세요.

# 미션 3-1: 연비(mpg)의 평균값과 표준편차
# cylinders로 그룹화하여 연비(mpg)의 평균값
df4 = df.groupby('cylinders')['mpg'].mean()
print(df4)

# cylinders로 그룹화하여 연비(mpg)의 표준편차
df5 = df.groupby('cylinders')['mpg'].std()
print(df5)

# origin을 그룹화하여 연비(mpg)의 평균값
df6 = df.groupby('origin')['mpg'].mean()
print(df6)

# origin을 그룹화하여 연비(mpg)의 표준편차
df7 = df.groupby('origin')['mpg'].std()
print(df7)
```

```python
# 미션 3-2: 마력(horsepower)의 최대값과 최소값
# cylinders로 그룹화하여 마력(horsepower)의 최대값
df8 = df2.groupby('cylinders')['horsepower'].max()
print(df8)

# cylinders로 그룹화하여 마력(horsepower)의 최소값
df9 = df2.groupby('cylinders')['horsepower'].min()
print(df9)

# origin을 그룹화하여 마력(horsepower)의 최대값
df10 = df2.groupby('origin')['horsepower'].max()
print(df10)

# origin을 그룹화하여 마력(horsepower)의 최소값
df11 = df2.groupby('origin')['horsepower'].min()
print(df11)
```
