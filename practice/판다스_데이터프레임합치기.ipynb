```python
## 데이터프레임 연결(concat)
# 라이브러리 불러오기
import pandas as pd

df1= pd.DataFrame({'a':['a0', 'a1', 'a2', 'a3'],
                 'b':['b0', 'b1', 'b2', 'b3'],
                 'c':['c0', 'c1', 'c2', 'c3']},
                index=[0,1,2,3])

df1
```

```python
df2 = pd.DataFrame({'a':['a2', 'a3', 'a4', 'a5'],
                 'b':['b2', 'b3', 'b4', 'b5'],
                 'c':['c2', 'c3', 'c4', 'c5'],
                 'd':['d2', 'd3', 'd4', 'd5']},
                index=[2,3,4,5])
df2
```

```python
# 2개의 데이터프레임을 위 아래 행 방향으로 이어 붙이듯 연결하기
result1= pd.concat([df1, df2])
result1
```

```python
# ignore_index=True 옵션 설정하기(인덱스를 다시 설정함)
result2 = pd.concat([df1, df2], ignore_index=True)
result2
```

```python
# join='inner' 옵션 적용하기(교집합)
result2_in = pd.concat([df1, df2], axis=0, join='inner')
result2_in
```

```python
# 2개의 프레임을 좌우 열 방향으로 이어 붙이듯 연결하기
result3 = pd.concat([df1, df2], axis= 1)
result3
```

```python
# join='inner' 옵션 적용하기(교집합)
result3_in = pd.concat([df1, df2], axis= 1, join='inner')
result3_in
```

```python
# 시리즈 만들기
sr1 = pd.Series(['e0', 'e1', 'e2', 'e3'], name = 'e')
sr2 = pd.Series(['f0', 'f1', 'f2'], name = 'f', index=[3,4,5])
sr3 = pd.Series(['g0', 'g1', 'g2', 'g3'], name = 'g')
sr4 = pd.Series(['a6', 'b6', 'c6', 'd6'], index=['a', 'b', 'c', 'd'])

# df1과 sr1을 좌우 열 방향으로 연결하기
result4=pd.concat([df1, sr1], axis=1)
result4
```

```python
# df2과 sr2을 좌우 열 방향으로 연결하기
result5=pd.concat([df2, sr2], axis=1)
result5
```

```python
# df2과 sr4를 위 아래 행 방향으로 연결하기
result6=pd.concat([df2, sr4.to_frame().T], ignore_index=True)
result6
```

```python
# sr1과 sr3을 좌우 열 방향으로 연결하기
result7=pd.concat([sr1, sr3], axis=1)
result7
```

```python
# sr1과 sr3을 위 아래ㅀ 연결하기
result8=pd.concat([sr1, sr3], axis=0)
result8
```

```python
## 데이터프레임 병합
# 데이터프레임 합치기(merge)

import pandas as pd

df1=pd.read_excel('./data/stock_price.xlsx')
df1
```

```python
df2=pd.read_excel('./data/stock_valuation.xlsx')
df2
```

```python
# 데이터프레임 합치기- 교집합
merge_inner = pd.merge(df1, df2, how='inner', on='id')
merge_inner
```

```python
# 데이터프레임 합치기- 합집합
merge_outer = pd.merge(df1, df2, how='outer', on='id')
merge_outer
```

```python
# 데이터프레임 합치기- 왼쪽 데이터프레임 기준, 키 값 분리
merge_left = pd.merge(df1, df2, how='left', left_on=['id', 'stock_name'], right_on=['id', 'name'])

merge_left
```

```python
# 데이터프레임 합치기- 오른쪽 데이터프레임 기준, 키 값 분리
merge_right = pd.merge(df1, df2, how='right', left_on=['id', 'stock_name'], right_on=['id', 'name'])

merge_right
```

```python
# 데이터프레임 합치기 - 교차조인(중복은 제거되지 않음)
merge_cross = pd.merge(df1, df2, how='cross')
merge_cross
```

```python
# 불린 인덱싱과 결합하여 원하는 데이터 찾기
price = df1[df1['price'] < 50000]
price
```

```python
value = pd.merge(price, df2)
value
```
