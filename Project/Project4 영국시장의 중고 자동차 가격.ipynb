```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df1 = pd.read_csv('./data/cars.csv')
df1.head()
```

```python
df1.info()
```

```python
df2 = pd.read_csv('./data/brand.csv')
df2.head()
```

```python
df2.info()
```

```python
# df1 title 데이터 확인
df1['title'].unique()
```

```python
# df1 title 데이터 확인
df2['title'].unique()
```

```python
# df1 title 데이터 슬라이스
df1.rename(columns={'title':'title_model'}, inplace = True)
```

```python
df1.head()
```

```python
df1[['title', 'model']]=df1['title_model'].str.split(' ', n=1, expand= True)
```

```python
df1.head()
```

```python
# df2 title 대문자로 변환
df2['title']=df2['title'].str.upper()
```

```python
df2.head()
```

```python
# cars.csv와 brand.csv의 데이터 합치기
df_join = pd.merge(df1, df2, on='title', how='left') 
```

```python
# 카테고리형 변수를 숫자 형태로 변환하기
df_join['Engine'] = df_join['Engine'].str.replace('L', '').astype(float)
df_join['Euro_Class'] = df_join['Emission Class'].str.extract(r'(\d+)').astype(float)
```

```python
df_join.head()
```

```python
# 데이터 개요 및 결측치/이상치 처리
df_join.describe()
```

```python
# 결측치 확인
df_join.isnull().sum()
```

```python
# Service history의 결측치 'Unknown'으로 채우기
df_join['Service history'] = df_join['Service history'].fillna('Unknown')
```

```python
# 결측치가 2개 이상인 행 제거
df_join =df_join[df_join.isnull().sum(axis=1) < 2]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 수치형 컬럼
numeric_cols = ['Price', 'Mileage(miles)', 'Registration_Year', 
                'Previous Owners', 'Engine', 'Doors', 'Seats', 'Euro_Class']

# 결측치 있는 행 제거
df_numeric = df_join[numeric_cols].dropna()
df_numeric.head()
```

```python
df_numeric.isnull().sum()
```

```python
df_numeric.describe()
```

```python
# 표준화 스케일링
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_numeric)
```

```python
# PCA(주성분) 분석
pca = PCA(n_components=0.70)
pca_data = pca.fit_transform(scaled_data)
```

```python
print(f"선택된 주성분 개수: {pca.n_components_}")
print(pca.explained_variance_ratio_)
```

```python
# 국가별 총 브랜드 개수
brand_counts = df_join.groupby('country')['title'].nunique().reset_index().rename(columns={'title': 'brand_count'})
brand_counts
```

```python
# 상관관계 분석 (heatmap)
corr_matrix = df_numeric.corr()
corr_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Correlation Heatmap of Numeric Features')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
